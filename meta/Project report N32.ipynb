## 2. Đề xuất project (W3)

### Bài toán
- Trong thời đại số hiện nay, âm nhạc đã trở thành một phần không thể thiếu trong đời sống con người. Mỗi người nghe nhạc đều có những sở thích riêng và mong muốn tìm được các bài hát phù hợp với tâm trạng, hoạt động hoặc gu thưởng thức của mình. Tuy nhiên, khối lượng bài hát trên các nền tảng nghe nhạc trực tuyến là vô cùng lớn, việc chọn lựa thủ công thường mất nhiều thời gian và dễ bỏ lỡ những gợi ý phù hợp.
- Vấn đề đặt ra là xây dựng một hệ thống gợi ý nhạc có khả năng đưa ra đề xuất bài hát hoặc thể loại nhạc phù hợp dựa trên một số thông tin đầu vào từ người dùng, chẳng hạn như tâm trạng (vui, buồn, thư giãn), hoạt động (học tập, tập thể dục) hay thể loại yêu thích (Pop, EDM, Rap, Cổ điển).
- Bài toán này có nhiều ứng dụng trong thực tiễn, từ việc cải thiện trải nghiệm cá nhân hóa trên các nền tảng nghe nhạc trực tuyến, đến hỗ trợ xây dựng playlist tự động cho sự kiện, hoạt động tập thể. Phương pháp giải quyết bài toán bao gồm biểu diễn tri thức dưới dạng luật và sử dụng cơ chế suy diễn logic để tìm ra tập gợi ý phù hợp với người dùng.

### Phương pháp
- Xây dựng cơ sở tri thức:
  - Tập hợp khoảng 30-40 bài hát mẫu, chia theo thể loại: EDM, Pop, Ballad, Rap, Cổ điển,…
  - Xây dựng tập luật if – then. Ví dụ:
    - If tâm trạng = vui → gợi ý EDM hoặc Pop, R&B.
    - If  tâm trạng = buồn → gợi ý Ballad hoặc Bolero.
    - If  hoạt động = tập thể dục → gợi ý Rap hoặc EDM.
    - If  thích thể loại ballad → gợi ý Hoà Minzy, Bảo Anh, Soobin,…
- Biểu diễn tri thức:
  - Biểu diễn luật bằng cấu trúc dữ liệu Python (dict, JSON).
  - Mỗi luật có 2 phần: conditions (điều kiện) và recommend (danh sách bài hát gợi ý).
- Cơ chế suy diễn:
  - Áp dụng suy diễn tiến (forward chaining):
    - Nhận input từ người dùng (ví dụ: “tâm trạng = vui”).
    - Kiểm tra các luật trong cơ sở tri thức.
    - Trả về danh sách gợi ý phù hợp.
- Cài đặt và thử nghiệm:
  - Ngôn ngữ: Python.
  - Tạo giao diện đơn giản (console hoặc web mini bằng Streamlit).
  - Thử nghiệm với nhiều trường hợp người dung nhập khác nhau.

## 3. Tiến độ giữa kỳ (W8)

### Chương trình
- Kiến trúc: Chương trình được xây dựng bằng Python với thư viện Streamlit cho giao diện web. Cấu trúc gồm:
  - Thư mục code: chứa toàn bộ code của chương trình
    - app.py: File chính chạy giao diện người dùng
    - inferenceEngine.py: Bộ não suy diễn, chứa thuật toán gợi ý.
    - recommender.py: Lớp điều phối chính, giao tiếp giữa giao diện và bộ não suy diễn.
  - Thư mục data: Chứa cơ sở tri thức và dữ liệu.
    - songs.json: Lưu trữ dữ liệu (Data) về 100 bài hát, bao gồm các thuộc tính như id, title, artist, genre, moods, activity và url.
    - rules.json: Biểu diễn tri thức dưới dạng Luật sản xuất gồm condition (điều kiện đơn dựa trên input người dùng) và effect (tác động cộng/trừ điểm cho thể loại/nghệ sĩ).
- Cơ chế gợi ý:
  - Áp dụng phương pháp Suy diễn tiến dựa trên điểm số (Tác tử dựa trên lợi ích).
  - Giai đoạn 1: Đối sánh trực tiếp mood/activity/genre từ input người dùng với thông tin trong songs.json, cộng điểm ưu tiên cao (+35) nếu khớp.
  - Giai đoạn 2: Kích hoạt các luật trong rules.json dựa trên tam_trang, hoat_dong, the_loai_yeu_thich của người dùng để cộng/trừ điểm cho các bài hát theo the_loai hoặc nghe_si.
  - Kết quả: Tổng hợp điểm từ hai giai đoạn, dùng hàm sorted() để xếp hạng và trả về top 5 bài hát có điểm dương cao nhất.

### Phân tích kết quả, vấn đề gặp phải
- Phân tích kết quả:
  - Hệ thống gợi ý thành công top 5 bài hát dựa trên input, hiển thị tên bài - điểm số.
  - Đã thử nghiệm với các input khác nhau (ví dụ: chỉ chọn tâm trạng, chỉ chọn hoạt động, kết hợp nhiều yếu tố) và cho ra kết quả có vẻ hợp lý ban đầu.
  - Giao diện Streamlit đã hoạt động, cho phép người dùng chọn Tâm trạng, Hoạt động, Thể loại yêu thích.
- Vấn đề gặp phải:
  - Bùng nổ tổ hợp luật: Ban đầu dự định tạo luật cho mọi kết hợp input (mood x activity x genre) -> Số lượng luật quá lớn (ước tính ~35k) -> Giải pháp: Chuyển sang mô hình luật đơn kết hợp tính điểm (score-based), giảm đáng kể số lượng luật cần viết (~ vài trăm) và tăng tính linh hoạt.
  - Với số lượng bài hát nhỏ, cách tiếp cận này đủ nhanh. Tuy nhiên, nếu songs.json được mở rộng lên hàng nghìn hoặc chục nghìn bài hát trong tương lai, thời gian phản hồi của hệ thống có thể sẽ chậm đi đáng kể do độ phức tạp tính toán là O(số_luật * số_bài_hát).
  - Vấn đề về giao diện người dùng: Giao diện web được xây dựng bằng thư viện Streamlit. Do còn đang trong quá trình làm quen và tìm hiểu các thành phần giao diện của Streamlit, nên giao diện hiện tại có thể chưa hoàn toàn tối ưu hoặc khớp chính xác 100% với bản phác thảo ban đầu. Tuy nhiên, các chức năng nhập liệu chính (chọn tâm trạng, hoạt động, thể loại) và hiển thị kết quả gợi ý đã được thực hiện. Việc hoàn thiện giao diện sẽ là một phần công việc trong giai đoạn tiếp theo.
- Dự định đến cuối kỳ:
  - Hoàn thiện và tinh chỉnh điểm số trong rules.json để cải thiện chất lượng gợi ý.
  - Bổ sung thêm các luật đơn còn thiếu cho các mood/activity chưa có.
  - Xem xét thêm các luật kết hợp AND cho các trường hợp đặc biệt (nếu cần).
  - Cân nhắc ứng dụng hệ chuyên gia vào project (sử dụng thư viện Experta).
