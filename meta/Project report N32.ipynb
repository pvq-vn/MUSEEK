## 2. Đề xuất project (W3)

### Bài toán
- Trong thời đại số hiện nay, âm nhạc đã trở thành một phần không thể thiếu trong đời sống con người. Mỗi người nghe nhạc đều có những sở thích riêng và mong muốn tìm được các bài hát phù hợp với tâm trạng, hoạt động hoặc gu thưởng thức của mình. Tuy nhiên, khối lượng bài hát trên các nền tảng nghe nhạc trực tuyến là vô cùng lớn, việc chọn lựa thủ công thường mất nhiều thời gian và dễ bỏ lỡ những gợi ý phù hợp.
- Vấn đề đặt ra là xây dựng một hệ thống gợi ý nhạc có khả năng đưa ra đề xuất bài hát hoặc thể loại nhạc phù hợp dựa trên một số thông tin đầu vào từ người dùng, chẳng hạn như tâm trạng (vui, buồn, thư giãn), hoạt động (học tập, tập thể dục) hay thể loại yêu thích (Pop, EDM, Rap, Cổ điển).
- Bài toán này có nhiều ứng dụng trong thực tiễn, từ việc cải thiện trải nghiệm cá nhân hóa trên các nền tảng nghe nhạc trực tuyến, đến hỗ trợ xây dựng playlist tự động cho sự kiện, hoạt động tập thể. Phương pháp giải quyết bài toán bao gồm biểu diễn tri thức dưới dạng luật và sử dụng cơ chế suy diễn logic để tìm ra tập gợi ý phù hợp với người dùng.

  - Mỗi luật có 2 phần: conditions (điều kiện) và recommend (danh sách bài hát gợi ý).
- Cơ chế suy diễn:
  - Áp dụng suy diễn tiến (forward chaining):
    - Nhận input từ người dùng (ví dụ: “tâm trạng = vui”).
    - Kiểm tra các luật trong cơ sở tri thức.
    - Trả về danh sách gợi ý phù hợp.
- Cài đặt và thử nghiệm:
  - Ngôn ngữ: Python.
  - Tạo giao diện đơn giản (console hoặc web mini bằng Streamlit).

## 4. Cập nhật kết quả cuối kỳ (W15)

### Chi tiết phương pháp, dữ liệu 

#### 1. Cách tiếp cận bài toán

- Hệ thống được xây dựng theo mô hình **Hệ chuyên gia dựa trên luật (Rule-Based Expert System)**. Thay vì sử dụng các thuật toán học máy (Machine Learning) hay lọc cộng tác (Collaborative Filtering) cần dữ liệu lịch sử lớn, nhóm sử dụng tri thức chuyên gia được mã hóa thành các luật dẫn để suy diễn ra những bài hát phù hợp với người dùng.

#### 2. Mô hình biểu diễn tri thức
Tri thức của hệ thống được chia làm hai phần:

- **Sự kiện (Facts)**: Là thông tin đầu vào từ người dùng (tâm trạng, hoạt động, thể loại yêu thích) và thông tin của các bài hát trong cơ sở dữ liệu.

- **Luật dẫn (Production Rules)**: Có dạng IF `<Condition>` THEN `<Action>`.
  - **Condition (Điều kiện)**: Các tổ hợp của tâm trạng, hoạt động, thể loại.
  - **Action (Hành động)**: Tăng/giảm điểm số trọng số (score) cho các thể loại nhạc hoặc nghệ sĩ cụ thể.

#### 3. Cơ chế suy diễn
Hệ thống sử dụng cơ chế **Suy diễn tiến (Forward Chaining)**.

- **Lý do chọn:** Bài toán đi từ dữ liệu có sẵn (input của người dùng) để tìm ra kết luận (danh sách bài hát phù hợp).

- **Quy trình:**
  - Nhận Input (Tâm trạng, hoạt động, thể loại yêu thích).
  - Hệ thống quét qua tập luật (rules.json).
  - Kích hoạt các luật thỏa mãn điều kiện (Matching).
  - Cập nhật "Bộ nhớ làm việc" (Working Memory) là danh sách điểm số ưu tiên của các thể loại (Genre) và nghệ sĩ (Artist).
  - Tính toán điểm số cuối cùng cho từng bài hát và ranking.

#### 4. Hàm mục tiêu

Độ phù hợp của một bài hát $S$ được tính toán dựa trên công thức trọng số:
  $$Score(S) = \sum Score_{direct} + \sum Score_{rule\_genre} + \sum Score_{rule\_artist}$$
Trong đó:
  - $Score_{direct}$: Điểm cộng trực tiếp (+50) nếu bài hát có metadata (mood, activity, genre) trùng khớp hoàn toàn với input.
  - $Score_{rule\_genre}$: Tổng điểm thưởng/phạt của các thể loại bài hát $S$ sở hữu, dựa trên các luật đã kích hoạt.
  - $Score_{rule\_artist}$: Tổng điểm thưởng/phạt của nghệ sĩ thể hiện bài hát $S$, dựa trên các luật đã kích hoạt.

#### 3. Kiến trúc chương trình
Hệ thống được viết bằng Python, cấu trúc theo mô hình phân tách logic:

##### a. app.py

- Chức năng: Đóng vai trò là tầng giao diện người dùng (Frontend), chịu trách nhiệm thu thập dữ liệu đầu vào (Input) và hiển thị kết quả gợi ý (Output). Sử dụng thư viện Streamlit.

- Cơ chế hoạt động & Logic:
  - Quản lý trạng thái (State Management): Sử dụng st.session_state để lưu trữ dữ liệu phiên làm việc (như danh sách bài hát gợi ý recommendations, bài hát đang phát active_media) nhằm tránh việc dữ liệu bị mất khi giao diện render lại sau mỗi tương tác người dùng.
  - Tối ưu hiệu năng (Caching): Sử dụng Decorator @st.cache_resource để áp dụng kỹ thuật Memoization. Hệ thống chỉ khởi tạo đối tượng Recommender và nạp dữ liệu từ ổ cứng một lần duy nhất khi server khởi động, giúp tăng tốc độ tải trang.
  - Xử lý sự kiện (Event Handling): Sử dụng cơ chế Callback trong các nút bấm (Button) để xử lý logic (VD: on_click=set_active_media) trước khi giao diện được vẽ lại.

##### b. recommender.py

- Chức năng: Đóng vai trò lớp điều khiển trung gian (Controller/Facade), quản lý dữ liệu và điều hướng yêu cầu.

- Cơ chế hoạt động:
  - Data Loading: Khởi tạo lớp Recommender, nạp dữ liệu từ songs.json và rules.json vào bộ nhớ RAM ngay khi ứng dụng khởi chạy.
  - Routing (Điều hướng): Hàm suggest() nhận đầu vào và ID của Engine, sau đó gọi đến hàm xử lý tương ứng (Inference/Rule/Experta).

##### c. inferenceEngine.py

- Chức năng: Cơ chế suy diễn thủ công, tự cài đặt để hiểu bản chất hệ chuyên gia.

- Thuật toán: Duyệt tuyến tính (Linear Search) kết hợp Suy diễn tiến (Forward Chaining).

- Logic xử lý:
  - Duyệt luật: Vòng lặp for quét qua toàn bộ danh sách rules.
  - Matching: So sánh chính xác (==) từng cặp key-value trong điều kiện của luật với input người dùng.
  - Hệ quả: Nếu khớp, cập nhật điểm số vào dictionary preferred_genres hoặc preferred_artists.
  - Ranking: Duyệt qua danh sách bài hát, tính tổng điểm = Điểm khớp trực tiếp (50đ) + Điểm trọng số từ luật. Sau đó sort giảm dần.

##### d. ruleEngine.py

- Chức năng: Sử dụng thư viện rule_engine của Python để xử lý logic mờ và cú pháp phức tạp hơn.

- Thuật toán: Abstract Syntax Tree (AST) Parsing.

- Logic xử lý:
  - Parsing: Chuyển đổi điều kiện trong JSON (Dictionary) thành chuỗi biểu thức logic Python (ví dụ: tam_trang == "Vui vẻ" and hoat_dong == "Party").
  - Filtering: Sử dụng rule_engine.Rule(rule_string) để tạo đối tượng luật và gọi phương thức .matches() để kiểm tra input.
  - Ranking: Tương tự như inferenceEngine.py.

##### e. expertaEngine.py

- Chức năng: Hệ chuyên gia hoàn chỉnh sử dụng thư viện experta (Python port của CLIPS).

- Thuật toán: Rete Algorithm. Đây là thuật toán pattern matching hiệu quả cao, giúp hệ thống không bị chậm đi tuyến tính khi số lượng luật tăng lên hàng nghìn.

- Logic xử lý:
  - Dynamic Rule Generation (Sinh luật động): Không viết cứng luật (hard-code). Chương trình đọc file rules.json và tự động sinh ra các hàm luật (@Rule) tại thời điểm chạy (Runtime) thông qua hàm create_dynamic_rule.
  - Fact Declaration: Chuyển input người dùng thành các sự kiện (UserContext(Fact)) và nạp vào Engine.
  - Firing Rules: Engine chạy (engine.run()), thuật toán Rete kích hoạt các luật phù hợp và bắn ra các Recommendation(Fact).
  - Aggregation: Một luật thu thập (collect_score) sẽ lắng nghe các Recommendation được bắn ra để tổng hợp điểm số cuối cùng.

#### 4. Đầu vào và đầu ra của chương trình

- Input: Đầu vào của hệ thống bao gồm hai thành phần chính:
  - Dữ liệu tri giác từ người dùng (User Percepts): Là các thông tin ngữ cảnh được thu thập thông qua giao diện (Sensors):
    - Tâm trạng ($M$): $M \in \{ \text{Vui vẻ, Buồn, Sôi động, Thư giãn, ...} \}$
    - Hoạt động ($A$): $A \in \{ \text{Lái xe, Hẹn hò, Party, Học tập, ...} \}$
    - Thể loại ưu tiên ($G_{pref}$): $G_{pref} \in \{ \text{Pop, Ballad, EDM, ...} \}$
    - Định dạng: Chuỗi ký tự (String), chọn từ danh sách cố định (SelectBox).
  - Cơ sở tri thức (Knowledge Base):
    - Tập dữ liệu bài hát ($S$): Chứa 120 bài hát với các thuộc tính định danh (Title, Artist) và thuộc tính phân loại (Moods, Activity, Genre).
    - Tập luật suy diễn ($R$): Tập hợp các luật dẫn (If-Then rules) trong file rules.json dùng để ánh xạ từ Input người dùng sang điểm số.

- Output: Hệ thống trả về danh sách các Hành động khuyến nghị (Recommended Actions), cụ thể:
  - Danh sách bài hát xếp hạng ($L_{ranked}$): Top 10 bài hát có tổng điểm độ thỏa dụng (Utility Score) cao nhất, sắp xếp giảm dần.
    - $Score(S) = \sum Score_{direct} + \sum Score_{rule\_genre} + \sum Score_{rule\_artist}$
  - Thông tin chi tiết (Metadata):
    - Tên bài hát, Nghệ sĩ trình bày.
  - Tác vụ đa phương tiện (Actuators):
    - Liên kết phát nhạc trực tiếp (YouTube Embed / Spotify Link).
    - Liên kết Karaoke (nếu có).

#### 2. Đánh giá hiệu năng và so sánh các cơ chế suy diễn

##### a. Phân tích độ phức tạp (Theoretical Complexity)

- Engine 1: InferenceEngine
  - Cơ chế: Duyệt tuần tự qua danh sách luật (for rule in rules), bên trong mỗi luật duyệt qua từng điều kiện (for key in conditions).
  - Độ phức tạp thời gian: $O(R \times C)$, trong đó $R$ là số lượng luật và $C$ là số điều kiện trung bình mỗi luật.
  - Nhận xét: Hiệu năng giảm tuyến tính khi số lượng luật tăng. Phù hợp với tập dữ liệu nhỏ (< 500 luật).
  - Ưu điểm: Kiểm soát hoàn toàn logic, không phụ thuộc thư viện bên ngoài (Zero-dependency), dễ debug từng dòng lệnh.
  - Nhược điểm: Hard-code logic so sánh (chỉ so sánh bằng ==), khó mở rộng sang các so sánh phức tạp (lớn hơn, nhỏ hơn, regex).

- Engine 2: RuleEngine (Python Library)
  - Cơ chế: Sử dụng kỹ thuật Abstract Syntax Tree (AST) để parse chuỗi điều kiện (ví dụ: "tam_trang == 'Vui vẻ'") thành cây logic.
  - Độ phức tạp: Phụ thuộc vào độ dài của chuỗi biểu thức logic. Việc parse string tốn chi phí CPU đáng kể ở thời điểm chạy (runtime) nếu không được cache.
  - Ưu điểm: Cú pháp viết luật rất linh hoạt, gần gũi với ngôn ngữ tự nhiên hoặc SQL. Tách biệt hoàn toàn dữ liệu luật (lưu trong DB/File text) khỏi mã nguồn.
  - Nhược điểm: Tốn tài nguyên CPU để parse chuỗi string thành lệnh thực thi mỗi lần chạy.

- Engine 3: ExpertaEngine (Rete Algorithm)
  - Cơ chế: Dựa trên thuật toán Rete (tương tự CLIPS). Thuật toán này xây dựng một mạng lưới (network) các nút điều kiện để tối ưu hoá việc khớp mẫu.
  - Độ phức tạp và Hiệu năng thực tế:
    - Lý thuyết: Thuật toán Rete có độ phức tạp khớp mẫu gần như $O(1)$ hoặc $O(\log N)$, cực kỳ nhanh khi mạng lưới đã được biên dịch.
    - Thực tế triển khai: Để đảm bảo tính linh hoạt (cho phép cập nhật file rules.json mà không cần sửa code nguồn), hệ thống sử dụng kỹ thuật Sinh luật động (Dynamic Rule Generation) tại runtime. Điều này khiến Engine phải khởi tạo và biên dịch lại mạng lưới Rete cho mỗi yêu cầu tìm kiếm (Request), tạo ra chi phí khởi tạo (overhead) nhất định. Do đó, tốc độ thực tế sẽ chậm hơn lý thuyết, nhưng đổi lại hệ thống đạt được tính bảo trì và mở rộng dữ liệu cao (Data-driven).
  - Ưu điểm: Là chuẩn mực của Hệ chuyên gia. Hỗ trợ suy diễn phức tạp, quản lý xung đột luật (Conflict Resolution), và tối ưu hóa cực tốt cho tập luật khổng lồ (hàng nghìn/triệu luật).
  - Nhược điểm: Cấu hình phức tạp (cần khai báo Fact, Schema), tiêu tốn nhiều RAM để dựng mạng lưới node ban đầu.

#### 2. Đánh giá ưu/nhược điểm

- Ưu điểm:
  - Tính giải thích được (Explainability): Có thể truy vết chính xác tại sao một bài hát được gợi ý (do luật nào kích hoạt).
  - Linh hoạt: Dễ dàng thêm luật mới vào rules.json mà không cần sửa code (Data-driven).
  - Đa dạng thuật toán: Có thể tự cài đặt hoặc dùng thư viện.

- Nhược điểm:
  - Dữ liệu tĩnh: Cần cập nhật thủ công file JSON khi có bài hát mới.
  - Xử lý ngôn ngữ tự nhiên: Input đang chọn từ list (SelectBox), chưa xử lý được câu văn tự do của người dùng.
  - Khớp chính xác (Hard matching): Các luật hiện tại so sánh chính xác chuỗi (string equality), chưa có Logic mờ (Fuzzy Logic) để xử lý các mức độ như "hơi buồn" hay "rất vui".
